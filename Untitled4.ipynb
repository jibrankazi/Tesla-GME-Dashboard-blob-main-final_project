```python
import yfinance as yf

# Create Ticker object for Tesla
tesla = yf.Ticker("TSLA")

# Get historical stock data
tesla_data = tesla.history(period="max")

# Reset index so 'Date' is a column, not the index
tesla_data.reset_index(inplace=True)

# Display first 5 rows
tesla_data.head()
```

```python
import yfinance as yf
import pandas as pd
# Create Tesla ticker object
tesla = yf.Ticker("TSLA")
# Get quarterly financial data
tesla_financials = tesla.quarterly_financials
# Extract revenue row and transpose
tesla_revenue = tesla_financials.loc["Total Revenue"].T
# Reset index for clean DataFrame
tesla_revenue = tesla_revenue.reset_index()
tesla_revenue.columns = ["Date", "Revenue"]
# Display result
tesla_revenue.head()
```

```python
import yfinance as yf

# Create Ticker object for GameStop
gme = yf.Ticker("GME")

# Get historical stock data
gme_data = gme.history(period="max")

# Reset index so 'Date' is a column
gme_data.reset_index(inplace=True)

# Show first 5 rows
gme_data.head()
```

```python
import yfinance as yf
import pandas as pd
# Create Ticker object for GameStop
gme = yf.Ticker("GME")
# Get quarterly financial data
gme_financials = gme.quarterly_financials
# Extract "Total Revenue" row and transpose it
gme_revenue = gme_financials.loc["Total Revenue"].T
# Reset index and rename columns
gme_revenue = gme_revenue.reset_index()
gme_revenue.columns = ["Date", "Revenue"]
# Display result
gme_revenue.head()
```

```python
import plotly.graph_objects as go
# Tesla Stock Price Plot
fig1 = go.Figure()
fig1.add_trace(go.Scatter(x=tesla_data["Date"], y=tesla_data["Close"], name="Stock Price"))
fig1.update_layout(
    title="Tesla Stock Price Over Time",
    xaxis_title="Date",
    yaxis_title="Price (USD)",
    width=800,     # smaller width
    height=400     # smaller height
)
fig1.show()
```

```python
# Tesla Revenue Plot
fig2 = go.Figure()
fig2.add_trace(go.Scatter(x=tesla_revenue["Date"], y=tesla_revenue["Revenue"], name="Revenue"))
fig2.update_layout(
    title="Tesla Revenue Over Time",
    xaxis_title="Date",
    yaxis_title="Revenue (USD)",
    width=800,
    height=400
)
fig2.show()
```

```python
import plotly.graph_objects as go
# GameStop Stock Price Plot
fig3 = go.Figure()
fig3.add_trace(go.Scatter(x=gme_data["Date"], y=gme_data["Close"], name="Stock Price"))
fig3.update_layout(
    title="GameStop Stock Price Over Time",
    xaxis_title="Date",
    yaxis_title="Price (USD)",
    width=800,
    height=400
)
fig3.show()
```

```python
# GameStop Revenue Plot
fig4 = go.Figure()
fig4.add_trace(go.Scatter(x=gme_revenue["Date"], y=gme_revenue["Revenue"], name="Revenue"))
fig4.update_layout(
    title="GameStop Revenue Over Time",
    xaxis_title="Date",
    yaxis_title="Revenue (USD)",
    width=800,
    height=400
)
fig4.show()
```
